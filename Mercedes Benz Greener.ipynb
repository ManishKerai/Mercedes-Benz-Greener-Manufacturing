```python
#separating y from the dataset

y_train = df_train['y'].values
```

```python
# iterating  through all the columns which has X in the name of the column

cols = [c for c in df_train.columns if 'X' in c]
```

```python
print('Number of features: {}'.format(len(cols)))
```

```python
df_train[cols].dtypes.value_counts()
```

```python
# Counting the data in each of the columns

counts = [[], [], []]
for c in cols:
    typ = df_train[c].dtype
    uniq = len(np.unique(df_train[c]))
    if uniq == 1:
        counts[0].append(c)
    elif uniq == 2 and typ == np.int64:
        counts[1].append(c)
    else:
        counts[2].append(c)
```

```python
print('Constant features: {} Binary features: {} Categorical features: {}\n'
      .format(*[len(c) for c in counts]))
print('Constant features:', counts[0])
print('Categorical features:', counts[2])
```

```python
#Reading test.csv file

df_test = pd.read_csv('test.csv')
```

```python
df_test.columns
```

```python
#removing columns ID and Y from the dataset

new_cols = list(set(df_train.columns) - set(['ID', 'y']))
```

```python
y_train = df_train['y'].values
```

```python
id_test = df_test['ID'].values
```

```python
X_train = df_train[new_cols]
X_test = df_test[new_cols]
```

```python
# checking for null values

def check_missing_values(df):
    if df.isnull().any().any():
        print("There are missing values in the dataframe")
    else:
        print("There are no missing values in the dataframe")
```

```python
check_missing_values(X_train)
```

```python
check_missing_values(X_test)
```

```python
#applying label encoder

for column in new_cols:
    cardinality = len(np.unique(X_train[column]))
    if cardinality == 1:
        X_train.drop(column, axis=1) # Column with only one 
        # value is useless so we drop it
        X_test.drop(column, axis=1)
    if cardinality > 2: # Column is categorical
        mapper = lambda x: sum([ord(digit) for digit in x])
        
        X_train[column] = X_train[column].apply(mapper)
        X_test[column] = X_test[column].apply(mapper)
```

### As we have large no. of features we must chose only the relavant ones for better accuarcy. We can opt for dimetionality reduction for this.

```python
n_comp = 12
pca = PCA(n_components= n_comp, random_state=101)
```

```python
pca2_results_train = pca.fit_transform(X_train)
```

```python
pca2_results_test = pca.transform(X_test)
```

```python
# Training using xgboost

import xgboost as xgb

from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(pca2_results_train, y_train, test_size=0.2, random_state=101)
```

```python
d_train = xgb.DMatrix(X_train, label=y_train)
```

```python
d_valid = xgb.DMatrix(X_valid, label=y_valid)
```

```python
d_test = xgb.DMatrix(pca2_results_test)
```

```python
params = {}
params['objective'] = 'reg:linear'
params['eta'] = 0.02
params['max_depth'] = 4
```

```python
def xgb_r2_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2', r2_score(labels, preds)
```

```python
watchlist = [(d_train, 'train'), (d_valid, 'valid')]
```

```python
clf = xgb.train(params, d_train, 
                1000, watchlist, early_stopping_rounds=50, 
                custom_metric=xgb_r2_score, maximize=True, evals=10)
```

```python
# Predicting  test_df values using xgboost

p_test = clf.predict(d_test)

sub = pd.DataFrame()
sub['ID'] = id_test
sub['y'] = p_test
sub.to_csv('xgb.csv', index=False)

sub.head()
```
